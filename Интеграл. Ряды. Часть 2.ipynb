### Исследовать ряд на сходимость, используя признак д’Аламбера: 
$$ \sum_{n=1}^{\infty}\frac{n^n}{(n!)^2}$$ Признак д’Аламбера выглядит следующим образом: $$ \lim_{n\to\infty}\frac{a_{n+1}}{a_n}=q,$$ при q<1 ряд сходится, при q>1 ряд расходится, при q=1 нужны дополнительные исследования. $$ q=\lim_{n\to\infty}\frac{a_{n+1}}{a_n}=\lim_{n\to\infty}\frac{(n+1)^{n+1}\cdot(n!)^2}{((n+1)!)^2\cdot n^n}=\lim_{n\to\infty}\frac{(n+1)^n\cdot(n+1)\cdot(n!)^2}{(n!\cdot(n+1))^2\cdot n^n}=\lim_{n\to\infty}\frac{(n+1)^n\cdot(n+1)\cdot(n!)^2}{(n!)^2\cdot(n+1)^2\cdot n^n}$$
Числитель и знаменатель сократим на $(n!)^2\cdot(n+1)$ и останется у нас: $$ q=\lim_{n\to\infty}\frac{(n+1)^n}{(n+1)\cdot n^n}=\lim_{n\to\infty}\frac{1}{n+1}\cdot\frac{(n+1)^n}{n^n}=\lim_{n\to\infty}\frac{1}{n+1}\cdot \lim_{n\to\infty}\frac{(n+1)^n}{n^n}$$ Рассмотрим поближе второй множитель: $$ \lim_{n\to\infty}\frac{(n+1)^n}{n^n}$$ Немного его преобразуем: $$ \lim_{n\to\infty}\frac{(n+1)^n}{n^n}=\lim_{n\to\infty}\left(\frac{n+1}{n}\right)^n=\lim_{n\to\infty}\left(\frac{n}{n}+\frac{1}{n}\right)^n=\lim_{n\to\infty}\left(1+\frac{1}{n}\right)^n=e,$$ После преобразований второй множитель превратился во второй замечательный предел! Таким образом у нас получилось: $$ \lim_{n\to\infty}\frac{1}{n+1}\cdot \lim_{n\to\infty}\frac{(n+1)^n}{n^n}=\lim_{n\to\infty}\frac{1}{n+1}\cdot e=\lim_{n\to\infty}\frac{e}{n+1}=0,$$ так как в числителе дроби константа (е - число Эйлера), а знаменатель в пределе стремится к бесконечности при $n\to\infty$
Таким образом q=0<1 $\implies$ ряд сходится.

### Исследовать ряд на сходимость, используя радикальный признак Коши:
$$ \sum_{n=1}^{\infty}\frac{n}{2^n}$$ Радикальный признак Коши выглядит так: $$ \lim_{n\to+\infty}\sqrt[n]{a_n}=q,$$ при q<1 ряд сходится, при q>1 ряд расходится, при q=1 радикальный признак Коши не работает. $$ q=\lim_{n\to+\infty}\sqrt[n]{\frac{n}{2^n}}=\frac{1}{2}\lim_{n\to+\infty}\sqrt[n]{n}=\frac{1}{2}\cdot 1=\frac{1}{2}$$ Таким образом q<1 $\implies$ ряд сходится.
(не могу объяснить почему $\sqrt[n]{n}\to 1$, при $n\to+\infty$ )

### Исследовать ряд на сходимость, используя признак Лейбница:
$$ \sum_{n=1}^\infty \frac{(-1)^n}{n+\ln n}$$ Решение по mathprofi.ru
Первым пунктом необходимо проверить, стремится ли к нулю при $n\to\infty$ модуль общего член ряда: $$ \lim_{n\to\infty}\left|\frac{(-1)^n}{n+\ln n}\right|=\frac{1}{n+\ln n}=\frac{1}{\infty}=0$$ Вторым пунктом проверим, является ли каждый последующий член ряда по модулю меньше, чем предыдущий по модулю. Много слов, математически все проще: $$ |a_n|\geq|a_{n+1}|$$ Для этого просто уберем (-) перед единицей: $$ \frac{1^n}{n+\ln n}>\frac{1^n}{n+1+\ln (n+1)}$$ Данное неравенство является истинным, так как в числителе единица, а с ростом n значение в знаменателе увеличивается, соответственно дробь справа от знака неравенства меньше дроби слева. Итак, при исследовании знакочередующегося ряда на сходимость по признаку Лейбница мы получили 1) общий член ряда по модулю стремится к нулю при $n\to\infty$, т.е. $\sum_{n=1}^\infty \left|\frac{(-1)^n}{n+\ln n}\right|$ ; 2) модуль каждого последующего члена ряда меньше модуля предыдущего члена ряда (опять много слов), т.е. ряд из модулей является сходящимся. Таким образом оба условия выполняются и ряд, указанный в задании, сходится абсолютно.

### Исследовать ряд на сходимость, используя признак Раабе: 
$$ \sum_{n=1}^\infty \frac{3^n}{2^n}$$ Признак Раабе гласит: если существует предел $$ \lim_{n\to\infty}\left(n\cdot\left(\frac{a_n}{a_{n+1}}-1\right)\right)=D,$$ то при D<1 ряд расходится, при D>1 ряд сходится, при D=1 признак не дает ответа.
(закатываю рукава) Приступим: $$ D=\lim_{n\to\infty}\left(n\cdot\left(\frac{3^n\cdot2^{n+1}}{2^n\cdot3^{n+1}}-1\right)\right)=\lim_{n\to\infty}\left(n\cdot\left(\frac{3^n\cdot2^n\cdot2}{2^n\cdot3^n\cdot3}-1\right)\right)=$$ После сокращений и прочей математической магии получим: $$ =\lim_{n\to\infty}-\frac{1}{3}n=-\infty$$ Итак, D<1, значит ряд расходится

### Разложить функцию по Тейлору в единице:
$$f(x)=\ln(16x^2)$$ Для начала вспомним, как раскладывается функция в ряд Тейлора: $$ f(x)=\sum_{n=0}^\infty\frac{f^{(n)}(a)}{n!}(x-a)^n$$ Разложение функции примет вид: $$ f(x)=\ln16 + 2(x-1) - (x-1)^2 + \frac{2}{3}(x-1)^3-\frac{1}{2}(x-1)^4+\frac{2}{5}(x-1)^5+...+\frac{f^{(n)}(1)}{n!}(x-1)^n$$ Обощив слагаемые под знак суммы получим разложение в следующе кратком виде: $$ f(x)=\ln16+\sum_{n=1}^{\infty}(-1)^{n+1}\frac{2}{n}(x-1)^n$$ Такая запись легко читаема и интуитивно понятна, но немного громоздка и ее можно упростить, если объединить множители с одинаковыми степенями: $$ f(x)=\ln16+\sum_{n=1}^{\infty}\frac{-2(1-x)^n}{n}$$ Такая запись кратка и математически красива, но логика вычисления каждого слагаемого немного спрятана внутри. Для вычисления вручную я бы оставил первый вариант, а для составления программы второй.
